```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.metrics import accuracy_score,classification_report
import pickle
```

```python
data=pd.read_csv("Churn_Modelling.csv")
print(data.head(5))
```

```python
data.info()
```

```python
data.describe()
```

```python
data.head(5)
```

```python
data.drop(["RowNumber","CustomerId","Surname"],axis=1,inplace=True)
```

```python
lb=LabelEncoder()
data["Gender"]=lb.fit_transform(data["Gender"])
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
geo=ohe.fit_transform(data[["Geography"]])
```

```python
geo=geo.toarray()
```

```python
ohe.get_feature_names_out()
```

```python
encoded=pd.DataFrame(geo,columns=ohe.get_feature_names_out())
```

```python
encoded
```

```python
data=pd.concat([data.drop("Geography",axis=1),encoded],axis=1)
```

```python
data.head(5)
```

```python
data.info()
```

```python
with open("label_encoder_geo.pkl","wb") as file:
    pickle.dump(lb,file)
```

```python
with open("OnehotEncoder.pkl","wb") as file:
    pickle.dump(ohe,file)
```

```python
x=data.drop(["Exited"],axis=1)
y=data[['Exited']]
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
with open("Scaler.pkl","wb") as file:
    pickle.dump(scaler,file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense 
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard 
import datetime 
```

```python
model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)), ##first Hidden layer 
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
opt=tf.keras.optimizers.legacy.Adam(learning_rate=0.01)
losss=tf.keras.losses.BinaryCrossentropy()
```

```python
model.compile(optimizer=opt,loss=losss,metrics=['accuracy'])
```

```python
log='logs/fit/'+datetime.datetime.now().strftime("%Y:%m:%d-%H:%M:%S")
callback=TensorBoard(log_dir=log,histogram_freq=1)
early=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
#training the model

history=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,
                  callbacks=[callback,early])
```

```python
model.save("mode.h5")
```
